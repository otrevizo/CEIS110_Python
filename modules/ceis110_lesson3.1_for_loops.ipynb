## First, do an example WITHOUT FOR loops

```python
s = 9
print('Next wind speed will be ', s, 'MPH')
if s > 10:
    print('It will be windy this hour')
else:
    print('It will be nice')
```

```python
s = 11
print('Next wind speed will be ', s, 'MPH')
if s > 10:
    print('It will be windy this hour')
else:
    print('It will be nice')
```

```python
s = 10
print('Next wind speed will be ', s, 'MPH')
if s > 10:
    print('It will be windy this hour')
else:
    print('It will be nice')
```

```python
s = 12
print('Next wind speed will be ', s, 'MPH')
if s > 10:
    print('It will be windy this hour')
else:
    print('It will be nice')
```

## Now use a FOR loop

```python
for s in [9, 11, 10, 12]:
    print('Next wind speed will be ', s, 'MPH')
    if s > 10:
        print('It will be windy this hour')
    else:
        print('It will be nice')
```

```python
f = [9, 11, 10, 12]
```

```python
# f is an iterator. So I can iterate in a loop.
for s in f:
    print('Next wind speed will be ', s, 'MPH')
    if s > 10:
        print('It will be windy this hour')
    else:
        print('It will be nice')
```
