```python
# Let's build a DataFrame.
# To do that, we need data. the data that will be contained in the DataFrame columns

first_name = ["Joan", "Mary", "Vijay ", "Rob",
              "Martha", "Josh", "Vicky", "Mario", "Jenny", "Joe"]
last_name = ["T", "K ", "N ", "R ", "L", "F ", "R", "L", "T", "P"]
score_1 = [91, 83, 95, 72, 91, 85, 89, 82, 80, 79]
score_2 = [91, 85, 90, 81, 95, 92, 88, 94, 78, 75]

classroom = {'First Name': first_name, 'Last Name': last_name,
              'Score 1': score_1, 'Score 2': score_2}
```

```python
type(classroom)
```

```python
# Now I can build a DataFrame.
# https://pandas.pydata.org/docs/reference/frame.html
    
df= pd.DataFrame(classroom)
```

```python
type(df)
```

```python
# DataFrame properties, attributes, and methods
dir(df)
```
