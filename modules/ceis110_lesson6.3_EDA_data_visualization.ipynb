## Grammar of Graphics (Leland Wilkinson)

Grammar of Graphics (GGPLOT from "R" computer language) is characterized by the following 7 elements:

1. Data: The information used to create the visualization. the foundation of the plot.
1. Aesthetics: Visual properties like color, size, shape, opacity, etc. that represent different aspects of the data.
1. Geometry: The basic visual to represent the data. Examples include points, lines, bars, areas, etc.
1. Facets: A way to arrange multiple plots within a single visualization, typically based on a categorical variable.
1. Statistics: Transformations applied to the data before visual representation.
1. Coordinates: The system that defines the position of visual marks on the plot.
1. Themes: A collection of visual properties that govern the overall look and feel of the plots.

In Python we will be using Matplotlib and or Seaborn. Python methods do not map  to the Grammar of Graphics framework in R GGPLOT does. Nevertheless, we will be able to think of our plots in the same way: We need data, will determine aesthetics, grometry, facets, statitistics, coordinates, and themes.

# Data prep: Filter for United, American and Delta Airlines in July

```python
# Filter the flights DataFrame. We will reduce its size right away
df = flights[(flights['carrier'].isin(['UA', 'AA', 'DL'])) & (flights['month'] == 7)]
df.info()
df.head()
```

# Histograms

## Histogram basics

1. Data: The newly filtered DataRame "df".
2. Aesthetics: The visual properties of the data that we will map to the plot elements: The x and y coordinates, color, etc... In this example "x" is "air_time".
3. Geometry: The alpha = 0.5 (opacity) and bins.
 

```python
# Plot histogram
plt.hist(df['air_time'], 
         bins=range(0, int(max(df['air_time'])) + 20, 20), 
         alpha=0.5, 
         color='blue', 
         edgecolor='black')

# Add labels and title
plt.xlabel('Air Time')
plt.ylabel('Frequency')
plt.title('Histogram of Air Time')
plt.grid(True)

# Show plot
plt.show()
```

## Histograms facets (with Matplotlib only)

Airports are JFK, EWR, and LGA 

```python
# Histogram example 2 with matplotlib only
# Plot histogram with facet_wrap using Matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Filter and plot for each origin
for i, origin in enumerate(df['origin'].unique()):
    ax = axes[i]
    ax.hist(df[df['origin'] == origin]['air_time'].dropna(), 
            bins=range(0, int(max(df['air_time'].dropna())) + 20, 20), 
            alpha=0.5, color='blue', edgecolor='black')
    ax.set_xlabel('Air Time')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Air Time for {origin}')
    ax.grid(True)

# Add a common title
fig.suptitle('Air time by NYC metro airport', fontsize=16, y=1.05)
plt.tight_layout()

# Show plot
plt.show()
```

## Histograms facets (with Seaborn)

### Hstograms with frequency

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, col='origin', col_wrap=3)

g.map(sns.histplot, 
      'air_time',
      bins=range(0, 600, 20), 
      stat='count', 
      kde=True, 
      color='blue', edgecolor='black', 
      alpha=0.5)

# Calculate the mean for each origin group
mean_values = df.groupby('origin')['air_time'].mean()

# Add vertical lines for the mean of each histogram
for ax, mean_value in zip(g.axes.flat, mean_values):
    ax.axvline(x=mean_value, color='red', linestyle='--', linewidth=1.5)

# Add titles and labels
g.set_titles('{col_name}')
g.set_axis_labels('Air time', 'Frequency')

# Show the plot
plt.show()
```

### Histograms with density

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, col='origin', col_wrap=3)

g.map(sns.histplot, 
      'air_time',
      bins=range(0, 600, 20), 
      stat='density', 
      kde=True, 
      color='blue', edgecolor='black', 
      alpha=0.5)

# Calculate the mean for each origin group
mean_values = df.groupby('origin')['air_time'].mean()

# Add vertical lines for the mean of each histogram
for ax, mean_value in zip(g.axes.flat, mean_values):
    ax.axvline(x=mean_value, color='red', linestyle='--', linewidth=1.5)

# Add titles and labels
g.set_titles('{col_name}')
g.set_axis_labels('Air time', 'Density')

# Show the plot
plt.show()
```
