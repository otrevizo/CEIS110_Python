## Grammar of Graphics (Leland Wilkinson)

Grammar of Graphics (GGPLOT from "R" computer language) is characterized by the following 7 elements:

1. Data: The information used to create the visualization. the foundation of the plot.
1. Aesthetics: Visual properties like color, size, shape, opacity, etc. that represent different aspects of the data.
1. Geometry: The basic visual to represent the data. Examples include points, lines, bars, areas, etc.
1. Facets: A way to arrange multiple plots within a single visualization, typically based on a categorical variable.
1. Statistics: Transformations applied to the data before visual representation.
1. Coordinates: The system that defines the position of visual marks on the plot.
1. Themes: A collection of visual properties that govern the overall look and feel of the plots.

In Python we will be using Matplotlib and or Seaborn. Python methods do not map  to the Grammar of Graphics framework in R GGPLOT does. Nevertheless, we will be able to think of our plots in the same way: We need data, will determine aesthetics, grometry, facets, statitistics, coordinates, and themes.

# Data prep: Filter for United, American and Delta Airlines in July

```python
# Filter the flights DataFrame. We will reduce its size right away
df = flights[(flights['carrier'].isin(['UA', 'AA', 'DL'])) & (flights['month'] == 7)]
df.info()
df.head()
```

# Histograms

## Histogram basics

1. Data: The newly filtered DataRame "df".
2. Aesthetics: The visual properties of the data that we will map to the plot elements: The x and y coordinates, color, etc... In this example "x" is "air_time".
3. Geometry: The alpha = 0.5 (opacity) and bins.
 

```python
# Plot histogram
plt.hist(df['air_time'], 
         bins=range(0, int(max(df['air_time'])) + 20, 20), 
         alpha=0.5, 
         color='blue', 
         edgecolor='black')

# Add labels and title
plt.xlabel('Air Time')
plt.ylabel('Frequency')
plt.title('Histogram of Air Time')
plt.grid(True)

# Show plot
plt.show()
```

## Histograms facets (with Matplotlib only)

Airports are JFK, EWR, and LGA 

```python
# Histogram example 2 with matplotlib only
# Plot histogram with facet_wrap using Matplotlib
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

# Filter and plot for each origin
for i, origin in enumerate(df['origin'].unique()):
    ax = axes[i]
    ax.hist(df[df['origin'] == origin]['air_time'].dropna(), 
            bins=range(0, int(max(df['air_time'].dropna())) + 20, 20), 
            alpha=0.5, color='blue', edgecolor='black')
    ax.set_xlabel('Air Time')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Histogram of Air Time for {origin}')
    ax.grid(True)

# Add a common title
fig.suptitle('Air time by NYC metro airport', fontsize=16, y=1.05)
plt.tight_layout()

# Show plot
plt.show()
```

## Histograms facets (with Seaborn)

### Histograms with frequency

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, col='origin', col_wrap=3)

g.map(sns.histplot, 
      'air_time',
      bins=range(0, 600, 20), 
      stat='count', 
      kde=True, 
      color='blue', edgecolor='black', 
      alpha=0.5)

# Calculate the mean for each origin group
mean_values = df.groupby('origin')['air_time'].mean()

# Add vertical lines for the mean of each histogram
for ax, mean_value in zip(g.axes.flat, mean_values):
    ax.axvline(x=mean_value, color='red', linestyle='--', linewidth=1.5)

# Add titles and labels
g.set_titles('{col_name}')
g.set_axis_labels('Air time', 'Frequency')

# Show the plot
plt.show()
```

### Histograms with density

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, col='origin', col_wrap=3)

g.map(sns.histplot, 
      'air_time',
      bins=range(0, 600, 20), 
      stat='density', 
      kde=True, 
      color='blue', edgecolor='black', 
      alpha=0.5)

# Calculate the mean for each origin group
mean_values = df.groupby('origin')['air_time'].mean()

# Add vertical lines for the mean of each histogram
for ax, mean_value in zip(g.axes.flat, mean_values):
    ax.axvline(x=mean_value, color='red', linestyle='--', linewidth=1.5)

# Add titles and labels
g.set_titles('{col_name}')
g.set_axis_labels('Air time', 'Density')

# Show the plot
plt.show()
```

### Histograms in matrix FacetGrid

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, 
                  row='origin', 
                  col='carrier', 
                  hue='origin', 
                  palette='Set1', 
                  legend_out=True)

g.map(sns.histplot, 
      'air_time',
      bins=range(0, 600, 20), 
      stat='count',
      edgecolor='black', 
      alpha=0.5)

# Add titles, labels, and adjust layout
g.fig.suptitle('Air time by NYC metro airport and carrier\nSource: transtats.bts.gov')
g.set_axis_labels('Air time', 'Number of flights')

# Adjust figure size and margins. Reduce margins slightly
g.fig.subplots_adjust(left=0.12, right=0.88, top=0.9, bottom=0.15)

# Show the plot
plt.show()
```

# Density plots (KDE)

Very similar to histograms. Kernel Density Estimate (KDE) plots do not plot each indivudual data point. The plots are based on a specific function ("Kernel") to estimate the data point based on a continuous line ("Density Estimate").

```python
# Create the FacetGrid with Seaborn
g = sns.FacetGrid(df, col='origin')

# Now the geometry is Density. Map the density plot to each subplot
g.map(sns.kdeplot, 
      'air_time', 
      fill=True, 
      alpha=0.5, 
      color='blue')

plt.suptitle('Air time by NYC metro airport\nSource: transtats.bts.gov', y=1.1)
g.fig.subplots_adjust(left=0.12, right=0.88, top=0.9, bottom=0.15)

plt.show()
```

# Boxplots

```python
# Create the boxplots with Seaborn's FacetGrid
g = sns.FacetGrid(df, col='origin')  # Adjust col_wrap if needed adding col_wrap = 3

g.map(sns.boxplot, 
      'carrier', 
      'air_time', 
      palette='Set1', 
      order=['AA', 'UA', 'DL'])  

g.fig.suptitle('Air Time by Carrier and Origin\nSource: transtats.bts.gov', y=1.1)
g.fig.subplots_adjust(left=0.12, right=0.88, top=0.9, bottom=0.15)

plt.show()
```

```python
# Create the boxplots with Seaborn's FacetGrid
g = sns.FacetGrid(df, col='origin')  # Adjust col_wrap if needed adding col_wrap = 3

g.map(sns.boxplot, 
      'carrier', 
      'air_time', 
      showmeans=False,
      showfliers=False,
      palette='Set1', 
      order=['AA', 'UA', 'DL'])  

g.fig.suptitle('Air Time by Carrier and Origin', y = 1.1)
g.fig.subplots_adjust(left=0.12, right=0.88, top=0.9, bottom=0.15)

plt.show()
```

# Scatterplots

## Example 1 - With regression line - Regular scale

```python
# Create scatter plot with Seaborn
g = sns.scatterplot(data=df, x='distance', y='air_time', hue='carrier')

# Add linear regression line
sns.regplot(data=df, x='distance', y='air_time', scatter=False, color='red')

# Add title and labels
plt.title('Air time vs distance')
plt.xlabel('Distance (miles)')
plt.ylabel('Air time (HHMM for hours:minutes)')

# Set x-axis limits
plt.xlim(0, 5100)

# Show the plot
plt.show()
```

## Example 2 - With regression line - Log scale

```python
# Create scatter plot with Seaborn
g = sns.scatterplot(data=df, x='distance', y='air_time', hue='carrier')

# Add linear regression line
sns.regplot(data=df, x='distance', y='air_time', scatter=False, color='red')

# Add title and labels
plt.title('Air time vs distance')
plt.xlabel('Log of distance (log miles)')
plt.ylabel('Log of air time (log of HHMM for hours:minutes)')

# Set x and y axis to log scale
plt.xscale('log')
plt.yscale('log')

# Adjust x- and y-axis tick marks
plt.xticks([100, 1000, 10000, 10000]) 
plt.yticks([10, 100, 1000]) 

# Show the plot
plt.show()
```

# Barplot

## Example 1

```python
# Wrangle your data
# Group, summarize, and arrange
df_grpby_origin = (df.groupby(['origin', 'carrier'])
                   .agg(
                       AVG_air_time=('air_time', 'mean'),
                       SD_air_time=('air_time', 'std'),
                       AVG_distance=('distance', 'mean'),
                       SD_distance=('distance', 'std'),
                       No_of_flights=('origin', 'count'),
                       TOTAL_distance=('distance', 'sum'))
                   .sort_values(by='AVG_air_time', ascending=False))

# Define custom order for carriers
carrier_order = ['AA', 'UA', 'DL']

# Quick exploration (optional)
print(df_grpby_origin.info())
print(df_grpby_origin)
```

```python
# Create the barplot: Number of fligths
sns.catplot(kind='bar',
            data=df_grpby_origin.reset_index(),
            x='origin',
            y='No_of_flights',
            hue='origin',  # Use 'hue' for categorical coloring
            col='carrier',
            palette='Set1',  # Set color palette
            legend=False,  # Hide legend
            col_order=carrier_order)  # Hide legend
plt.show()
```

```python
# Create the barplot: Total distance
sns.catplot(kind='bar',
            data=df_grpby_origin.reset_index(),
            x='origin',
            y='TOTAL_distance',
            hue='origin',  # Use 'hue' for categorical coloring
            col='carrier',
            palette='Set1',  # Set color palette
            legend=False,  # Hide legend
            col_order=carrier_order)  # Hide legend
plt.show()
```

```python
# Create the barplot: Average distance
sns.catplot(kind='bar',
            data=df_grpby_origin.reset_index(),
            x='origin',
            y='AVG_distance',
            hue='origin',  # Use 'hue' for categorical coloring
            col='carrier',
            palette='Set1',  # Set color palette
            legend=False,  # Hide legend
            col_order=carrier_order)  # Hide legend
plt.show()
```

```python
# Create the barplot: Average air time
sns.catplot(kind='bar',
            data=df_grpby_origin.reset_index(),
            x='origin',
            y='AVG_air_time',
            hue='origin',  # Use 'hue' for categorical coloring
            col='carrier',
            palette='Set1',  # Set color palette
            legend=False,  # Hide legend
            col_order=carrier_order)  # Hide legend
plt.show()
```
